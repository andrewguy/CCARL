```python
from ccarl.glycan_plotting import *
```

```python
GLYCAN_CODES_MIN = [['Hexose', 'Glc', 'Man', 'Gal', 'Gul', 'Alt', 'All', 'Tal', 'Ido', ''],
 ['HexNAc',
  'GlcNAc',
  'ManNAc',
  'GalNAc',
  'GulNAc',
  'AltNAc',
  'AllNAc',
  'TalNAc',
  'IdoNAc',
  ''],
 ['Deoxyhexose', 'Qui', 'Rha', '', '6dGul', '6dAlt', '', '6dTal', '', 'Fuc'],
 ['Deoxynonulosonate',
  '',
  'Kdn',
  '',
  '',
  '',
  'Neu5Ac',
  'Neu5Gc',
  'Neu',
  'Sia ']]
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(nrows=len(GLYCAN_CODES_MIN), ncols=len(GLYCAN_CODES_MIN[0]), figsize=(15,7))
for i, glycan_row in enumerate(GLYCAN_CODES_MIN):
    for j, glycan in enumerate(glycan_row):
        if glycan:
            draw_glycan(glycan, ax[i][j], 0.5, 0.5)
        ax[i][j].set_xlim((0.2, 0.8))
        ax[i][j].set_aspect('equal')
        ax[i][j].text(0.5, 0.2, glycan, fontsize=16, horizontalalignment='center')
        ax[i][j].axis('off')
fig.tight_layout()
fig.savefig('glycan_codes.svg')
```
